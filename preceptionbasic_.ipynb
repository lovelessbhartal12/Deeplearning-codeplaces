```python
from sklearn.datasets import make_classification
import numpy as np

x,y=make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False ,class_sep=10)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
```

```python
import numpy as np

# Step activation function
def step(z):
    return 1 if z >= 0 else 0

def perceptron1(x, y):
    x = np.insert(x, 0, 1, axis=1)
    weights = np.ones(x.shape[1])
    lr = 0.01

    for i in range(1000):
        j = np.random.randint(0, len(x)) # Pick random sample
        y_hat = step(np.dot(x[j], weights))
        weights = weights + lr * (y[j] - y_hat) * x[j]

    return weights[0], weights[1:]
```

```python
intercept_ ,coef=perceptron1(x,y)
```

```python
print(intercept_)
print(coef)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_decision_boundary(X, Y, bias, weights):
    plt.figure(figsize=(6,6))

    # Scatter plot of classes
    plt.scatter(X[Y==0,0], X[Y==0,1], color="red", label="Class 0")
    plt.scatter(X[Y==1,0], X[Y==1,1], color="blue", label="Class 1")

    # Decision boundary: bias + w1*x1 + w2*x2 = 0
    x1 = np.linspace(X[:,0].min()-0.5, X[:,0].max()+0.5, 100)
    x2 = -(bias + weights[0]*x1) / weights[1]

    plt.plot(x1, x2, color="green", label="Decision Boundary")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("Perceptron Decision Boundary")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_decision_boundary(x,y,intercept_,coef)
```
